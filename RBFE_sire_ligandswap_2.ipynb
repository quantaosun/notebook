一个好的判断我们的计算结果是否可靠的简单方法是看一下，Bennetts， FEP还有TI的数据是否接近

根据上面的结果我们可以看出，最低的一个数据是4.47，最高的是4.80，差距非常的小，可以得出平均相对结合自由能在4.7kcal/mol 的结论。注意我们使用了三个方法，但是有四个数据，因为TI会产生两个数据，我们使用这四个数据进行平均，就可以得到一个我们感觉信心最高的数据4.7. 就误差而言，三个方法的平均误差为0.5左右，在1 以内，因此，也说明这个数据的可靠性。
